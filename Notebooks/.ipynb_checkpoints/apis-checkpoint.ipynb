```python
def get_ldo_historical_data():
    ldo_historical_api = "https://api.coingecko.com/api/v3/coins/lido-dao/market_chart?vs_currency=usd&days=1152&interval=daily"

    # Send a GET request to the API endpoint
    response10 = requests.get(ldo_historical_api)

    # Check for a valid response
    if response10.status_code == 200:
        # Parse the JSON response
        ldo_historical_pricedata = response10.json()

        # Extract the 'prices' and 'market_caps' lists from the JSON data
        ldo_historical_price = ldo_historical_pricedata['prices']
        ldo_market_cap = pd.DataFrame(ldo_historical_pricedata['market_caps'], columns=['date', 'marketcap'])
        ldo_history = pd.DataFrame(ldo_historical_price, columns=['date', 'price'])

        # Calculate supply
        ldo_supply = ldo_market_cap['marketcap'] / ldo_history['price']

        return ldo_history, ldo_market_cap, ldo_supply

# Use the function in your Streamlit app
ldo_history, ldo_market_cap, ldo_supply = get_ldo_historical_data()
```
